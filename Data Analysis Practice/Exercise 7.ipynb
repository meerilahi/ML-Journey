Data Cleaning with Pandas

```python
import numpy as np
import pandas as pd
```

#Dealing with Missing Values

##Indentifying Missing Values

```python
pd.isnull(np.nan)
```

```python
pd.isnull(None)
```

```python
pd.isna(np.nan)
```

```python
pd.isnull(pd.Series([1, np.nan, 7]))
```

```python
df = pd.DataFrame({
    'A': [1, np.nan, 7],
    'B': [np.nan, 2, 3],
    'C': [0, 2, 3],
    'D': [4, np.nan, np.nan],
    'E': [4, 8, 6]
})
df.isnull()
```

##Droping Null Values

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=1, how='all')
```

```python
df.dropna(axis=1, thresh=2)
```

##Filling Null Values

```python
df.fillna(0)
```

```python
df.fillna(df.mean())
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(axis = 1, method='ffill')
```

#Dealing with Invalid Data

```python
df = pd.DataFrame({
    'Sex' : ['M', 'F', 'F', 'D', '?'],
    'Age' : [29, 30, 24, 290, 25],
})
df
```

##Finding Unique Values

```python
df.Sex.unique()
```

```python
df['Sex'].value_counts()
```

```python
df['Sex'].replace('D', 'F')
```

```python
df.drop(df[df['Sex'] == '?'].index)
```

```python
df.loc[df['Age'] > 100, 'Age'] = df.loc[df['Age'] > 100, 'Age'] / 10
df
```

##Duplicates

```python
ambassadors = pd.Series([
    'France',
    'United Kingdom',
    'United Kingdom',
    'Italy',
    'Germany',
    'Germany',
    'Germany',
], index=[
    'Gérard Araud',
    'Kim Darroch',
    'Peter Westmacott',
    'Armando Varricchio',
    'Peter Wittig',
    'Peter Ammon',
    'Klaus Scharioth '
])
```

```python
ambassadors
```

```python
ambassadors.duplicated()
```

```python
ambassadors.duplicated(keep='last')
```

```python
ambassadors.duplicated(keep=False)
```

```python
ambassadors.drop_duplicates(keep='last')
```

```python
ambassadors[ambassadors.duplicated(keep=False)]
```

```python
players = pd.DataFrame({
    'Name': [
        'Kobe Bryant',
        'LeBron James',
        'Kobe Bryant',
        'Carmelo Anthony',
        'Kobe Bryant',
    ],
    'Pos': [
        'SG',
        'SF',
        'SG',
        'SF',
        'SF'
    ]
})
```

```python
players
```

```python
players.duplicated()
```

```python
players.duplicated(subset=['Name'], keep='last')
```

```python
players.drop_duplicates()
```

```python
players.drop_duplicates(subset=['Name'], keep='last')
```

##Text Handling

```python
df = pd.DataFrame({
    'Data': [
        '1987_M_US _1',
        '1990?_M_UK_1',
        '1992_F_US_2',
        '1970?_M_   IT_1',
        '1985_F_I  T_2'
]})
```

```python
df
```

```python
df['Data'].str.split('_')
```

```python
df = df['Data'].str.split('_', expand=True)
```

```python
df.columns = ['Year', 'Sex', 'Country', 'No Children']
```
