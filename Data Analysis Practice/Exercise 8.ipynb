Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Global API

```python
x = np.arange(-10,11)
```

```python
plt.figure(figsize = (12,6))
plt.plot(x, x**2)
plt.plot(x, -1*(x**2))
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.title = 'My First Plot'

plt.subplot(1,2,1)
plt.plot(x, x**2)
plt.plot([0,0,0],[-10,0,100])
plt.legend('X^2','Vertical Line')
plt.xlabel('X')
plt.ylabel('X Square')

plt.subplot(1,2,2)
plt.plot(x, -1*(x**2))
plt.plot([-10,0,10],[-50,-50,-50])
plt.legend('-X^2','Horizontal Line')

plt.xlabel('X')
plt.ylabel('X Square')
```

OOP API

```python
x = np.arange(1,10)
y1 = 1/(2**x)
y2 = 2**x
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y2)
axes.plot(x,y1)
fig.show()
```

```python
x = np.arange(-10,11)
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(nrows = 3, ncols = 3, figsize = (18,9))

ax1.set_title('Constant Function')
ax1.plot(x,x+2, c = 'yellow')

ax2.set_title('Linear Function')
ax2.plot(x,2*x, c = 'red')

ax3.set_title('Quadratic Function')
ax3.plot(x,x**2, c = 'green')

ax4.set_title('Cubic Function')
ax4.plot(x,x**3, c = 'blue')

ax5.set_title('Exponential Function')
ax5.plot(x[11:], 2**(x[11:]), c = 'pink')

ax6.set_title('Logarithmic Function')
ax6.plot(x, np.log(x), c = 'grey')

ax7.set_title('Sin Function')
ax7.plot(x, np.sin(x), c = 'orange')

ax8.set_title('Cosine Function')
ax8.plot(x, np.cos(x), c = 'purple')

ax9.set_title('Tangent Function')
ax9.plot(x, np.tan(x), c = 'brown')
```
