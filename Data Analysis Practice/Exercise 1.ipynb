# Correlation Analysis and Visualization

```python
corr = df.select_dtypes(include = ['number']).corr()
```

```python
fig = plt.figure(figsize = (10,5))
plt.matshow(corr,cmap='RdBu', fignum = fig.number)
plt.xticks(range(len(corr.columns)), corr.columns, rotation = 'vertical')
plt.yticks(range(len(corr.columns)), corr.columns)
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i, f"{corr.iloc[i, j]:.1f}", ha='center', va='center', color='black')
```

# Column Wrangling

Creating Revenue Per Age Column

```python
df['Revenue_per_age'] = df['Revenue'] / df['Customer_Age']
```

```python
fig = df['Revenue_per_age'].plot(kind = 'density')
fig.set_title('Revenue Per Age')
```

```python
df['Revenue_per_age'].plot(kind = 'hist')
```

Validating Cost Column

```python
df['Calculated_cost'] = df['Order_Quantity'] * df['Unit_Cost']
(df['Calculated_cost'] != df['Cost']).sum()
```

```python
df.plot(kind = 'scatter', x = 'Calculated_cost', y = 'Profit')
```

Validating Revenue Column

```python
df['Calculated_Revenue'] = df['Cost'] + df['Profit']
(df['Calculated_Revenue'] != df['Revenue']).sum()
```

```python
df['Calculated_Revenue'].plot(kind = 'hist')
```

Increasing Unit Prices by 3 percent

```python
df['Unit_Price'] *= 1.03
```

# Selection and Indexing

Get All Sales made in state of Kentucky

```python
df.loc[df['State'] == 'Kentucky']
```

Get the mean Revenue of Adults (35-64) age group

```python
df.loc[df['Age_Group'] == 'Adults (35-64)', 'Revenue'].mean()
```

How many records belong to Age Group Youth (<25) or Adults (35-64)

```python
((df['Age_Group'] == 'Youth (<25)') | (df['Age_Group'] == 'Adults (35-64)')).sum()
```

Get the mean revenue of age group Youth (<25) in United States

```python
df.loc[(df['Age_Group'] == 'Youth (<25)') & (df['Country'] == 'United States'), 'Revenue'].mean()
```

Increase the Revenue by 10% to every sale made in France

```python
df.loc[df['Country'] == 'France', 'Revenue'] *= 1.1
```
